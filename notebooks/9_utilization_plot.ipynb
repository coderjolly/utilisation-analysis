```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("7_plot.csv")
df = df.iloc[ : , 1: ]
df.columns = ["date", "time", "sim_util", "actual_sim_util"]
df
```

```python
days = set(df.date)
days
```

```python
def getChartForDay(dt: pd.core.frame.DataFrame, date: str):
    data = dt[dt.date == date]
    plt.figure(figsize=(16,5))
    plt.plot(data.time, data.sim_util, "-o")
    plt.plot(data.time, data.actual_sim_util, "-x")

    plt.title(f'{date} Day Plot')
    plt.xlabel('Hours')
    plt.ylabel('Percentage Utilization')
```

```python
for day in sorted(days):
    getChartForDay(df, day)
```
