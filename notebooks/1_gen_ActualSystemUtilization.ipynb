```python
import pandas as pd
```

```python
from datetime import datetime as dt
```

```python
act_df = pd.read_csv("3_apst_03_2021.csv",names=["Timestamp","Configured","Up","Reserved","Used","A","B","C","D","E"])
```

```python
act_df["Timestamp"] = pd.to_datetime(act_df["Timestamp"])
lower_limit_dt = dt(2021,3,8,21)
upper_limit_dt = dt(2021,3,22,21)
```

```python
sliced_act_df = act_df.loc[(act_df["Timestamp"]>=lower_limit_dt) & (act_df["Timestamp"]<=upper_limit_dt)]
```

```python
sliced_act_df
```

```python
sliced_act_df["Date"] = sliced_act_df["Timestamp"].dt.strftime("%Y/%m/%d")
```

```python
sliced_act_df
```

```python
sliced_act_df["Hours"] = sliced_act_df["Timestamp"].dt.strftime("%H")
```

```python
sliced_act_df
```

```python
act_util_df=sliced_act_df[["Date","Hours","Up","Reserved"]].groupby(["Date","Hours"]).mean()
```

```python
act_util_df["SystemUtilization"] = round(act_util_df["Reserved"]/act_util_df["Up"]*100,2)
```

```python
act_util_df.to_csv("5_ActualSystemUtilization.csv")
```
